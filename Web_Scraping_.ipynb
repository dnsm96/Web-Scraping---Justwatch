````python
#import all necessary labraries
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
import numpy as np
````

## **Scrapping Movies Data**

````python
# Specifying the URL from which movies related data will be fetched
url='https://www.justwatch.com/in/movies?release_year_from=2000'

# Sending an HTTP GET request to the URL
page=requests.get(url)
# Parsing the HTML content using BeautifulSoup with the 'html.parser'
soup=BeautifulSoup(page.text,'html.parser')
# Printing the prettified HTML content
print(soup.prettify())
````

## **Fetching Movie URL's**

````python
data = soup.find_all('div',class_='title-list-grid__item')
````

````python
url = []
for i in data:
  link = i.a.get('href')
  url.append(link)
print(url)
````

````python
active_url = []
for i in url:
  final_link = 'https://www.justwatch.com'+i
  active_url.append(final_link)
print(active_url)
print(len(active_url))
````

## **Scrapping Movie Title**

````python
# Write Your Code here
movie_title = []
for title in data:
  name = title.get('data-title')
  movie_title.append(name)
print(movie_title)
print(len(movie_title))
````

## **Scrapping release Year**

````python
# Write Your Code here
years = []
for url in active_url:
  w = requests.get(url)
  soup = BeautifulSoup(w.text,'html.parser')
  year_span = soup.find('span',class_='text-muted')

  if year_span:
    year_text = year_span.get_text(strip=True)
    year = year_text.split()[-1]
    year = ''.join(filter(str.isdigit,year))
    years.append(year)
  else:
    years.append(np.nan)

print(years)
print(len(years))
````

## **Scrapping Genres**

````python
# Write Your Code here
Geners = []
for i in active_url:
  page1 = requests.get(i)
  soup1 = BeautifulSoup(page1.text,'html.parser')
  try:
    gen = soup1.div.find('span',class_='detail-infos__value').text
    Geners.append(gen)
  except:
    Geners.append(np.nan)

````

````python
print(Geners)
print(len(Geners))
````

## **Scrapping IMBD Rating**

````python
# Write Your Code here
IMDB_Rating = []
for i in active_url:
  page1 = requests.get(i)
  soup1 = BeautifulSoup(page1.text,"html.parser")
  try:
    data = soup1.find_all('div',class_='jw-scoring-listing__rating')
    IMDB_Rating.append((data[-1].text.split())[0])
  except:
    IMDB_Rating.append(np.nan)
````

````python
print(IMDB_Rating)
print(len(IMDB_Rating))
````

## **Scrapping Runtime/Duration**

````python
# Write Your Code here
Runtime = []
for i in active_url:
  page1 = requests.get(i)
  soup1 = BeautifulSoup(page1.text,'html.parser')
  try:
    IMDBrat = soup1.find('h3', string='Runtime').find_next_sibling('div',class_='detail-infos__value')
    Runtime.append(IMDBrat.text)
  except:
   Runtime.append(np.nan)
````

````python
print(Runtime)
print(len(Runtime))
````

## **Scrapping Age Rating**

````python
# Write Your Code here
Age_rating = []
for i in active_url:
  page1 = requests.get(i)
  soup1 = BeautifulSoup(page1.text,'html.parser')
  try:
    IMDBrat = soup1.find('h3', string='Age rating').find_next_sibling('div',class_='detail-infos__value')
    Age_rating.append(IMDBrat.text)
  except:
    Age_rating.append(np.nan)
````

````python
print(Age_rating)
print(len(Age_rating))
````

## **Fetching Production Countries Details**

````python
# Write Your Code here
Production_country = []
for i in active_url:
  page1 = requests.get(i)
  soup1 = BeautifulSoup(page1.text,'html.parser')
  try:
    IMDBrat = soup1.find('h3', string=' Production country ').find_next_sibling('div',class_='detail-infos__value')
    Production_country.append(IMDBrat.text)
  except:
    Production_country.append(np.nan)
````

````python
print(Production_country)
print(len(Production_country))
````

## **Fetching Streaming Service Details**

````python
# Write Your Code here
# Write Your Code here
Streaming_Service_Details = []
for i in active_url:
  page1 = requests.get(i)
  soup1 = BeautifulSoup(page1.text,'html.parser')
  try:
    IMDBrat = soup1.find('div',class_='buybox-row__offers')
    img_tag = IMDBrat.find('img', class_='offer__icon')
    Streaming_Service_Details.append(img_tag.get('alt'))
  except:
    Streaming_Service_Details.append(np.nan)
````

````python
print(Streaming_Service_Details)
print(len(Streaming_Service_Details))
````

````python
# Write Your Code here
data = {
    'url' : active_url,
    'Title' : movie_title,
    'Release Year': years,
    'Genres':Geners,
    'IMDB_Rating':IMDB_Rating,
    'Runtime':Runtime,
    'Age_rating':Age_rating,
    'Production_country':Production_country,
    'Streaming_Service_Details':Streaming_Service_Details
    }

df = pd.DataFrame(data)
df
````
